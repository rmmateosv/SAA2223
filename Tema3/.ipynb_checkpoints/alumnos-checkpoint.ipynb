# 1 Prepocesado de datos<a name="1"></a>
[Índice](#indice)

https://archive.ics.uci.edu/ml/datasets/student+performance

1 school - student's school (binary: "GP" - Gabriel Pereira or "MS" - Mousinho da Silveira)

2 sex - student's sex (binary: "F" - female or "M" - male)

3 age - student's age (numeric: from 15 to 22)

4 address - student's home address type (binary: "U" - urban or "R" - rural)

5 famsize - family size (binary: "LE3" - less or equal to 3 or "GT3" - greater than 3)

6 Pstatus - parent's cohabitation status (binary: "T" - living together or "A" - apart)

7 Medu - mother's education (numeric: 0 - none,  1 - primary education (4th grade), 2 – 5th to 9th 
grade, 3 – secondary education or 4 – higher education)

8 Fedu - father's education (numeric: 0 - none,  1 - primary education (4th grade), 2 – 5th to 9th grade, 3 – secondary education or 4 – higher education)

9 Mjob - mother's job (nominal: "teacher", "health" care related, civil "services" (e.g. administrative or police), "at_home" or "other")

10 Fjob - father's job (nominal: "teacher", "health" care related, civil "services" (e.g. administrative or police), "at_home" or "other")

11 reason - reason to choose this school (nominal: close to "home", school "reputation", "course" preference or "other")

12 guardian - student's guardian (nominal: "mother", "father" or "other")

13 traveltime - home to school travel time (numeric: 1 - <15 min., 2 - 15 to 30 min., 3 - 30 min. to 1 hour, or 4 - >1 hour)

14 studytime - weekly study time (numeric: 1 - <2 hours, 2 - 2 to 5 hours, 3 - 5 to 10 hours, or 4 - >10 hours)

15 failures - number of past class failures (numeric: n if 1<=n<3, else 4)

16 schoolsup - extra educational support (binary: yes or no)

17 famsup - family educational support (binary: yes or no)

18 paid - extra paid classes within the course subject (Math or Portuguese) (binary: yes or no)

19 activities - extra-curricular activities (binary: yes or no)

20 nursery - attended nursery school (binary: yes or no)

21 higher - wants to take higher education (binary: yes or no)

22 internet - Internet access at home (binary: yes or no)

23 romantic - with a romantic relationship (binary: yes or no)

24 famrel - quality of family relationships (numeric: from 1 - very bad to 5 - excellent)

25 freetime - free time after school (numeric: from 1 - very low to 5 - very high)

26 goout - going out with friends (numeric: from 1 - very low to 5 - very high)

27 Dalc - workday alcohol consumption (numeric: from 1 - very low to 5 - very high)

28 Walc - weekend alcohol consumption (numeric: from 1 - very low to 5 - very high)

29 health - current health status (numeric: from 1 - very bad to 5 - very good)

30 absences - number of school absences (numeric: from 0 to 93)

31 G1 - first period grade (numeric: from 0 to 20)

31 G2 - second period grade (numeric: from 0 to 20)

32 G3 - final grade (numeric: from 0 to 20, output target)

### 2.1.2 Frecuencias<a name="2_1_2"></a>
[Índice](#indice)

```python
# Suele hacerse para variables object
df['school'].value_counts()
```

```python
# Idem con porcentajes
df['school'].value_counts()/len(df)*100
df['school'].value_counts()/df.shape[0]*100
```

```python
# Frecuencias de columna numérica
print(df.age.value_counts())
print(df.age.value_counts()/len(df)*100)
```

```python
# Idem ordenado por edad
print(df.age.value_counts().sort_index)
print(df.age.value_counts().sort_index()/len(df)*100)
```

```python
# Analizar frecuencias con gráficas
# Ver Nº de alumnos por tipo de escula
df.school.value_counts().plot(kind='bar', title='Nº de alumnos por tipo de escuela')
```

```python
# Analizar frecuencias con gráficas
# Ver Nº de alumnos por tipo de escula
df.school.value_counts().plot(kind='pie', autopct='%0.2f%%', title='Nº de alumnos por tipo de escuela')
```

```python
# Mostrar Nº de alumnos por edad
df['age'].value_counts().sort_index().plot(kind='bar', title='Nº de alumnos por edad', xlabel='Edad')
```

```python
# Gráficas con seaborn -> Hacer import!!
# Configuramos el tema
sns.set_theme(style='whitegrid')
ax=sns.catplot(x='age', kind='count', data=df)
ax.set(title='Nº de alumnos por edad')
```

```python
# Mostrar Nº de alumnos por edad diferenciando por sexo
ax=sns.catplot(x='age', kind='count', hue='sex', data=df)
ax.set(title='Nº de alumnos por edad')
```

```python
# Mostrar Nº de alumnos por edad diferenciando por sexo y tipo de escula
ax=sns.catplot(x='age', kind='count', hue='sex', row='school', data=df, aspect=2, margin_titles=True)
ax.set(title='Nº de alumnos por edad')
```

### 2.1.3 Agrupaciones<a name="2_1_3"></a>
[Índice](#indice)

```python
# Hacer grupos por variable/s
# Suelen hacerse para variables categóricas
# Obetener las estadísticas agrupando por sexo
df.groupby('sex').describe()
```

```python
# Estadísticas de una variables(G1,G2,G3) agrupado por sexo
df.groupby('sex')[['G1','G2','G3']].describe()
```

```python
# Media de notas por sexo
df.groupby('sex')[['G1','G2','G3']].mean()
```

```python
#GRupos de dos variables
#Medias de notas por sexo y escuela
df.groupby(['sex','school'])[['G1','G2','G3']].mean()
```

```python
# GRupos usando gráficas de seaborn (barplot)
#Obtenemos la nota media de G3 por sexo
g3porSexo=df.groupby('sex')['G3'].mean()
g3porSexo
```

```python
# Convertimos g3porSexo en un dataframe con índice
g3porSexo = g3porSexo.reset_index()
g3porSexo
```

```python
#Pintar gráfica
ax=sns.barplot(x='sex',y='G3',data=g3porSexo)
ax.set(title='Nota media del tercer trimestre por sexo')
```

```python
# Calculamos media
g3SexoEscuela = df.groupby(['sex','school'])['G3'].mean()
# Creamos df g3SexoEscuela
g3SexoEscuela =g3SexoEscuela.reset_index()

ax=sns.barplot(x='sex',y='G3', hue= 'school', data=g3SexoEscuela)
ax.set(title='Nota media del tercer trimestre por sexo y escuela')
```

### 2.1.4 Filtros<a name="2_1_4"></a>
[Índice](#indice)

```python
# Valores del campo guardian
df['guardian'].value_counts()
```

```python
# Obtener los datos donde la custodia no la tiene ni el padre ni la madre
filtro = df['guardian']=='other'
df[filtro]
#df[df['guardian']=='other']
```

```python
#Obtener notas (G1,G2 y G3) de los alumnos cuya custodia es other
df[filtro][['G1','G2','G3']]
```

```python
#Obtener la nota media de (G1,G2 y G3) de los alumnos cuya custodia es other
df[filtro][['G1','G2','G3']].mean()
```

```python
#Obtener la nota media de (G1,G2 y G3) de los alumnos cuya custodia es other por sexo
df[filtro].groupby('sex')[['G1','G2','G3']].mean()
```

```python
# Obtener el nº de alumnos que aprueban el tercer trimestre
filtroAprobados = df['G3']>=10
#Datos de alumnos aprobados
df[filtroAprobados]
#Nº de alumnos aprobados
len(df[filtroAprobados])
```

```python
# Función loc: Aplica un filtro y selecciona una columna
# Mostrar notas de alumnos aprobados G3 con función loc
print(f"Notas de alumnos aprobados\n{df.loc[filtroAprobados,'G3']}")
# Nota media de los alumnos aprobados G3
print(f"Notas media alumnos aprobados{df.loc[filtroAprobados,'G3'].mean()}")
```

```python
# Notas medias de los aprobados
df.loc[filtroAprobados,['G1','G2','G3']].mean()
```
