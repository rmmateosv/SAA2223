# 1 DT para todas las variables con df sin normalizar<a name="2"></a>
[Índice](#indice)

```python
# Separamos las variables independientes de la variable objetivo
x = df.columns
x = x.drop('Price')
x
```

```python
# Dividir en datos de entremnamiento y de test
x_train,x_test, y_train, y_test = train_test_split(df[x],df['Price'],train_size=0.80,random_state=42)
```

```python
# Crear el modelo con parámetros por defecto: No hay límite en cuanto a la profundidad, al menso una muesrta en 
# nodos hojas y al menos 2 muesrtas para crear un nodo 
modelo = DecisionTreeRegressor()
#entrenamos
modelo.fit(x_train,y_train)
```

```python
# Evaluar predicciones en train
y_train_predict = modelo.predict(x_train)
# Mostrar métricas
print(f"MSE TRAIN:{mean_squared_error(y_train,y_train_predict)}")
print(f"R2 TRAIN:{r2_score(y_train,y_train_predict)}")
```

```python
# Evaluar predicciones en test
y_test_predict = modelo.predict(x_test)
# Mostrar métricas
print(f"MSE TRAIN:{mean_squared_error(y_test,y_test_predict)}")
print(f"R2 TRAIN:{r2_score(y_test,y_test_predict)}")
```

```python
# Ver profundidad del árbol
modelo.get_depth()
```

```python
# Análisis del resultado: En test, es el mejor resultado de los 3 algoritmos que hemos visto. El resutlado de train es 
# demasiado perfecto ya que no hay error y el % de acierto es del 100% lo cual es un indicador de que hay sobreajuste, 
# es debido a que no hay restricciones en cuanto a la profunidad, nº de muestras en los nodos hoja y nº de muesras para
# crea nuevos nodos.
```

```python
# Crear el modelo con parámetros modificados
modelo = DecisionTreeRegressor(max_depth=10,min_samples_split=5,min_samples_leaf=5)
#entrenamos
modelo.fit(x_train,y_train)
```

```python
# Evaluar predicciones en train
y_train_predict = modelo.predict(x_train)
# Mostrar métricas
print(f"MSE TRAIN:{mean_squared_error(y_train,y_train_predict)}")
print(f"R2 TRAIN:{r2_score(y_train,y_train_predict)}")
```

```python
# Evaluar predicciones en test
y_test_predict = modelo.predict(x_test)
# Mostrar métricas
print(f"MSE TRAIN:{mean_squared_error(y_test,y_test_predict)}")
print(f"R2 TRAIN:{r2_score(y_test,y_test_predict)}")
```

```python
# Análisis del resultado: Hemos sido un poco más exigentes en los parámetros por lo que el resultado ha sido un poco peor
# tanto en train como en test, pero sigue siendo el mejor resultado de los tres modelos vistos hasta ahora.
```

# 4 DT con GridSearchCV <a name="4"></a>
[Índice](#indice)

```python
# Crear el diccionario con los valores de los parámetros con los que queremos entrenar
parametros ={
   'max_depth':[5,10,15],
    'min_samples_split':range(10,50)
}
```

```python
# Crear modelo
modelo = GridSearchCV(DecisionTreeRegressor(),parametros)
```

```python
%%time
# Entrenar modelo
modelo.fit(x_train,y_train)
```

```python
# ¿Cuál ha sido la mejor combinación de parámetros?
modelo.best_params_
```

```python
# Predecir en train
y_train_pred = modelo.best_estimator_.predict(x_train)
# Mostrar métricas
print(f"MSE TRAIN:{mean_squared_error(y_train,y_train_pred)}")
print(f"R2 TRAIN:{r2_score(y_train,y_train_pred)}")
```

```python
# Predecir en test
y_test_pred = modelo.best_estimator_.predict(x_test)
# Mostrar métricas
print(f"MSE TEST:{mean_squared_error(y_test,y_test_pred)}")
print(f"R2 TEST:{r2_score(y_test,y_test_pred)}")
```
