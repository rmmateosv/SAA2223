Descripción del dataset: El dataset contiene información de clientes de un banco y la variable objetivo es una variable binaria que refleja si el cliente cierra la cuenta o si continúa en el banco. Las variables son:
* _RowNumber_: Id de fila       
* _CustomerId_: Id del cliente      
* _Surname_: Apellidos          
* _CreditScore_: Puntuación que asigna el banco al cliente    
* _Geography_: País del cliente       
* _Gender_: Sexo
* _Age_: Edad del cliente             
* _Tenure_: Antigüedad del cliente en el banco, en años           
* _Balance_: Saldo del cliente          
* _NumOfProducts_: Nº de productos que el cliente tiene contratado en el banco   
* _HasCrCard_: 1 el cliente tiene tarjeta de crédito        
* _IsActiveMember_: 1 El cliente es un miembro activo  
* _EstimatedSalary_: Salario estimado en dólares  
* _Exited_: 0 el cliente no cierra la cuenta, 1 el cliente cierra la cuenta

```python
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

# Normalización
from sklearn.preprocessing import StandardScaler

# Modelo
from sklearn.svm import SVC

# Dividir tranin y test
from sklearn.model_selection import train_test_split, GridSearchCV

# Matriz de confusión
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Métricas
from sklearn.metrics import accuracy_score, precision_score, recall_score
from sklearn.metrics import f1_score, roc_auc_score, roc_curve, RocCurveDisplay
from sklearn.metrics import classification_report
```

```python
def metricas(yReal, yEstimada):
    print(f"Accuracy: {accuracy_score(yReal,yEstimada)}")
    print(f"Precision: {precision_score(yReal,yEstimada)}")
    print(f"Recall: {recall_score(yReal,yEstimada)}")
    print(f"AUC: {roc_auc_score(yReal,yEstimada)}")
```

## 3.2 Transformación de características <a name="3_2"></a>
[Índice](#indice)

```python
# Transforamcion de tipo
for c in ['HasCrCard','IsActiveMember','Exited']:
    df[c] = pd.Categorical(df[c])
```

```python
# Discretizar CreditScore (Bajo, Medio, Alto)
df['puntuacion'],cortes = pd.cut(df['CreditScore'],bins=3,labels=[0,1,2],retbins=True)
df[['puntuacion','CreditScore']]
```

```python
# Label Encoding
# No es necesario ya que Age y Tenure hemos dicho que las dejamos numéricas y puntuación ya lo tiene hecho
```

```python
# One hot encoding. Variables Geography y Gender
df = pd.get_dummies(df, columns=['Geography','Gender'], drop_first=True)
```

```python
# Cambiar a categóricas las columnas generadas
for c in df.select_dtypes('uint8'):
    df[c] = pd.Categorical(df[c])
```

## 3.3 Selección de características relevantes <a name="3_3"></a>
[Índice](#indice)

```python
# Borrar columnas irrelevantes
df.drop(['RowNumber','CustomerId','Surname','CreditScore'], axis=1,inplace=True)
```

# 4  SVM/SVC todas las variables con el DF sin tratar y parámetros por defecto <a name="4"></a>
[Índice](#indice)

```python
# Dividir datos de train y test
y = df['Exited']
x = df.drop(['Exited'],axis=1)

# Parámetro stratify: Mantiene la proporción de los valoares de la variable objetivo en y_train y en y_test
# Si en y tenemos un 80% de 0 y un 20% de 1, esa proporcion se va a mantener en y_train y en y_test
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.20, random_state=1, stratify=y)
```

```python
# Creamos y entranamos el modelo
modelo = SVC()
modelo.fit(x_train,y_train)
```

```python
# Mostrar la métrica por defecto del modelo, en este caso el accuracy
print(f"Accuracy en Train:{modelo.score(x_train,y_train)}")
print(f"Accuracy en Test:{modelo.score(x_test,y_test)}")
```

```python
# Predicción
y_pred = modelo.predict(x_train)
y_test_pred = modelo.predict(x_test)
```

```python
# Mostrar métricas
print("Métricas en train")
metricas(y_train,y_pred)
print("\nMétricas en test")
metricas(y_test,y_test_pred)
```

```python
# Las métricas reflejan un buenísimo resultado en train,  pero 
# un mal resultado en test => hay SOBREAJUSTE
```

# 7  SVM/SVC con GridSearchCV <a name="7"></a>
